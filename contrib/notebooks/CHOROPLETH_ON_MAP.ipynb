### <font color='steelblue'>Description</font>

Realizar mapa coroplético.
 
- Los datos para el mapa podran venir dados en un fichero shp, gojson, ....
- Estos datos traerán la geometrís más una serie de columnas cuya información está asociada a dichas geometrías
- La idea es representar esas geometrías con la información que se desee del resto de columnas

- Como parámetros fundamentales de deberá poder eligir:
    - La columna que condiciona el color de las coropletas
    - Las columns que contienen la información que se desea salga en la leyenda al seleccionar cada coropleta
    - El "fondo" del mapa sobre el que salen las coropletas se entiende que se puede elegir (Open Street Maps, ...). 
    
- Se aporta imagen a modo ejemplo de lo que se desea
- Se adjuntan varios sets de datos para prueba: barrios, distritos y teselas de Madrid

## <font color='green'>Settings</font>

```python
# Libraries to use
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

## <font color='green'>Data Load</font>

<font color='green'><b> TESELAS</b></font>

```python
fichero='./Datos/TESSELLATION_MADRID.geojson'
# Generamos un GEO DataFrame
CHOROPLETH = gpd.read_file(fichero) 
```

```python
CHOROPLETH
```

```python
#Definimos el nombre de la figura y los ejes que vamos a utilizar
fig, ax = plt.subplots(figsize=(10, 10))
#Colocamos el Título del mapa
ax.set_title('Testing', fontdict={'fontsize': '24', 'fontweight': '3'})
#Utilizamos la función plot de pandas para seleccionar la columna que queremos graficar y las características de formato en el mapa
CHOROPLETH.plot(cmap='Set3', ax=ax,zorder=5)
```
