## Klib

```python
# returns a visualization of the number and frequency of categorical features
klib.cat_plot(data)
```

```python
# returns a color-encoded correlation matrix
klib.corr_mat(data)
```

```python
# returns a color-encoded heatmap, ideal for correlations
klib.corr_plot(data)
```

```python
# displaying only positive correlations, other settings include threshold
klib.corr_plot(data, split='pos')
```

```python
# displaying only negative correlations
klib.corr_plot(data, split='neg')
```

```python
# default representation of correlations with the feature column
klib.corr_plot(data, target='left')
```

```python
# returns a distribution plot for every numeric feature
klib.dist_plot(data)
```

```python
# returns a figure containing information about missing values
klib.missingval_plot(data)
```

```python
# representation of the 4 most & least common values in each categorical column
klib.cat_plot(data, top=4, bottom=4) 
```
