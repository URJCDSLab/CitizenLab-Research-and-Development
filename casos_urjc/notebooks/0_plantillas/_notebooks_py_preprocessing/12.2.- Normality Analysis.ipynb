## Transformations

### Log Transform

<font color='darkgrey'><b>Operation</font>

```python
column_log=np.log(column)
column_log.skew()
```

```python
sns.histplot(data=data_num, x=column_log, kde=True)
```

### Square Root Transform

<font color='darkgrey'><b>Operation</font>

```python
column_sqrt=np.sqrt(column)
column_sqrt.skew()
```

```python
sns.histplot(data=data_num, x=column_sqrt, kde=True)
```

### Box-Cox Transform

<font color='darkgrey'><b>Operation</font>

```python
from scipy import stats
column_box=stats.boxcox(column)[0]
pd.Series(column_box).skew()
```

```python
sns.histplot(data=data_num, x=column_box, kde=True)
```
