## Data Visualization

### Line plot

```python
data.plot()
```

### Scatter plot

```python
''' INTRO: ESPECIFICAR LAS COLUMNAS QUE SE DESEAN'''
''' Los nombre de las columnas son case sensitive'''
data.plot.scatter(x='Plasma', y='Age')
```

### Area plot

```python
data.plot.area();
```

```python
''' INTRO: ESPECIFICAR LAS COLUMNAS QUE SE DESEAN'''
''' Los nombre de las columnas son case sensitive'''
columns = ['Plasma', 'Age']
data[columns].plot.area()
```

### Bar chart

```python
data.plot.bar();
```

```python
data.plot.bar(stacked=True);
```

```python
data.plot.barh();
```

```python
data.plot.barh(stacked=True)
```

```python
''' INTRO: ESPECIFICAR LAS COLUMNAS QUE SE DESEAN'''
''' Los nombre de las columnas son case sensitive'''
data.groupby('Age').mean().plot.bar()
```

### Piechart

```python
''' INTRO: ESPECIFICAR LAS COLUMNAS QUE SE DESEAN'''
''' Los nombre de las columnas son case sensitive'''
data.groupby('Plasma').count().plot.pie(y='Age')
```

### Histogram

```python
data.hist()
plt.tight_layout()
```

### Kernel density function

```python
x=data.plot.kde(subplots=True, figsize=(5,9))
```

```python
''' INTRO: ESPECIFICAR LA COLUMNA QUE SE DESEA'''
''' Los nombre de las columnas son case sensitive'''
sn.kdeplot(data['Plasma'], shade=True, color='orangered')
```

### Boxplot

```python
x=data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(20, 20), color='deeppink')
```

```python
''' INTRO: ESPECIFICAR LAS COLUMNAS QUE SE DESEAN'''
''' Los nombre de las columnas son case sensitive'''
columns = ['Plasma', 'Age']
data[columns].plot.box()
plt.xticks(rotation='vertical')
```

### Scatter matrix plot

```python
sn.pairplot(data)
```

```python
sn.scatterplot(x='Plasma', y='Age', data=data, hue='Age')
```
